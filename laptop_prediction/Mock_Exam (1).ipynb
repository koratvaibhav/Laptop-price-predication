# **Outliers**

 7. What techniques can be utilized to detect and eliminate any
 outliers present in the dataset?

```python
for i in num:
  plt.figure(figsize=(5,10))
  sns.boxplot(df[i])
  plt.show()
```

# outliers **Handaling**

```python
for i in num:
  Q1 = df[i].quantile(0.25)
  Q3 = df[i].quantile(0.75)
  IQr = Q3-Q1
  lower = Q1-1.5*IQr
  upper = Q3+1.5*IQr
  handle = df[(df[i]>=lower) & (df[i]<=upper)]
```

```python
df1 = handle.copy()
```
